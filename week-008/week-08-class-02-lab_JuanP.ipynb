# Exercise 2

Create a function that will take a list of web pages and return a list of all the anchor tags on those pages.

```python
# create functions
def get_links(url):
    from bs4 import BeautifulSoup
    import requests

#def get_links(url):
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text)
    a_tags = soup.find_all('a')
    links = []
    for anchor in a_tags:
        try:
            if (link := anchor['href']).startswith('http'):
                links.append(link)
            else:
                links.append(url + link)
        except KeyError:
            continue
    return links

get_links('https://www.nytimes.com') #r/learnpython
```

# Exercise 3:

Create a function that accesses a web page and returns data from that page.

```python

def web_scrapper(url):
    import requests
    from bs4 import BeautifulSoup

    url = url #"https://www.politico.com/2020-election/results/"

    r = requests.get(url)

    r.text

    soup = BeautifulSoup(r.text)

    tags = soup.findAll('div')

    for i in tags:
        try:
            if i['class'][1] == 'dem-count':
                h3 = i.find('h3')
                spans = h3.findAll('span')
                if 'Biden ' in spans[2].contents:
                    print(f"Biden has {spans[1].contents[0]} Electoral College votes")
            if i['class'][1] == 'gop-count':
                h3 = i.find('h3')
                spans = h3.findAll('span')
                if 'Trump ' in spans[0].contents:
                    print(f"Trump has {spans[1].contents[0]} Electoral College votes")
        except Exception:
            continue

web_scrapper("https://www.politico.com/2020-election/results/")
```
