```python
class SafeFloat:
    # replace pass with your code
    def __init__(self, number):
        self.num = number
        self.num_as_float = self.to_float() #calls to_float to set value
        
    def to_float(self):
        try:
            return(float(self.num)) #if error occurs during conversion, except block is executed
        except:
            return None #num_as_float gets set to None if num cannot be converted to float
    
    def check_type(self):
        try:
            if isinstance(self.num_as_float,float): #check if num_as_float is of float type
                return True
            else:
                return False #means num_as_float is of None Type
        except:
            #except block will be executed if num_as_float does not exist
            self.num_as_float = None 
            return False
    
    def __add__(self, other):
        if self.check_type() == True and other.check_type()==True:
            return self.num_as_float + other.num_as_float
        else:
            return 'Addition not supported' #if num_as_float is None type, then this msg will be shown

# test with int, bool
e = SafeFloat(3)
f = SafeFloat(False) # False is converted to 0 during float conversion
print(e.num_as_float)
print(f.num_as_float)
print(e.check_type())
print(f.check_type())
print(e + f)
print('')

# test with float, string
e = SafeFloat(3.5)
f = SafeFloat('Dave') # False is converted to 0 during float conversion
print(e.num_as_float)
print(f.num_as_float)
print(e.check_type())
print(f.check_type())
print(e + f)#unsupported addition
```
