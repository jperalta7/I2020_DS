# Exercise 1

Plot one chart using matplotlib (OOP-style) using the following data.

```python
df = pd.read_csv('../datasets/GOOG.csv')
df
```

```python
dfplot = df.iloc[:30].sort_values(by="Date")
x_axis = dfplot["Date"]
y_high = dfplot["High"]
y_low = dfplot["Low"]
plt.figure(figsize=(18,6))
plt.plot(x_axis, y_high, color='green')
plt.plot(x_axis, y_low, color='red')
plt.xlabel = x_axis

plt.show()
```
