# Exercise 2

Perform some initial exploratory analysis to the following dataframe. Then plot two or more charts related to the data.

```python
df_testers = pd.read_csv('../datasets/SHSAT-Registrations-and-Testers.csv')
df_testers
```

```python
df_testers.info()
```

```python
df_testers = df_testers.rename(columns={'Number of students who registered for the SHSAT': 'SAT Registered', "Number of students who took the SHSAT": "SAT Tested"})
```

```python
df_testers.info()
```

```python
sns.pairplot(df_testers)
plt.show()
```

```python
df_testers.corr()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(x=df_testers["SAT Registered"].astype(int), y=df_testers["SAT Tested"])

plt.ylabel('SAT Tested')
plt.xlabel('SAT Registered')
plt.show()
```

```python
# Heatmap showing average arrival delay for each airline by month
plt.title("Correlation Analysis")
sns.heatmap(data=df_testers[["Grade level", "SAT Registered", 'SAT Tested']].corr(), annot=True)

plt.show()
```

```python

#y_train.values.reshape(-1,1)
x = df_testers["SAT Registered"]
y = df_testers["SAT Tested"]
x = x.values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
# reg.fit(X_train.values.reshape(-1,1), y_train.values.reshape(-1,1))
reg.fit(X_train, y_train)
```

```python
# score = reg.score(X_test.values.reshape(-1,1), y_test.values.reshape(-1,1))
score = reg.score(X_test, y_test.values)
```

```python
fit_line = np.arange(45).reshape(-1,1)
sns.regplot(x="SAT Registered", y="SAT Tested", data=df_testers, fit_reg=True)
plt.plot(fit_line, reg.predict(fit_line))

plt.show()
```
